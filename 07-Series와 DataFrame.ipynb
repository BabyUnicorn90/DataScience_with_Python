```python
#Series: 여러개의 단일타입(scalar) 1차원 데이터   
#  --리스트와 다른 점: 단일타입의 데이터
import pandas as pd

s = pd.Series(["홍길동", 25])
print(s, type(s))
```

```python
#series의 index확인하기: .index
print(s.index)

#series생성시 index값을 임의로 부여할 수 있다. 
s2 = pd.Series(['홍길동', 25], index = ["Name", "Age"])
print(s2)
print(s2.index)

#이미 만들어진 시리즈의 인덱스를 추가 부여/수정할 수 있다.
s.index = ["Name", "Age"]
print(s)
print(s.index)
```

```python
kor = [80, 90, 75, 100, 65]                          #시리즈 데이터로 활용할 리스트 
print("리스트: ", kor)

print("\n")

kor_s = pd.Series(kor)                                    #시리즈 생성
print("시리즈: ", "\n", kor_s)

print("\n")

print("시리즈의 기초통계량 확인: ", "\n", kor_s.describe())     #기초통계량 확인: .describe()
```

```python
#기초 통계함수 
print(kor_s.min(), kor_s.max())      #최솟값, 최댓값
print(kor_s.mean(), kor_s.median())  #산술평균, 중앙값
print(kor_s.std())                   #표준편차

print(kor_s.isin([100]))             #kor_s에 100이 포함되어 있는지 확인하기
```

```python
#series연산
s3 = pd.Series([1, 5, 8, 4, 6, 1, 10])
print("s3: ", "\n", s3)

print("\n")

#series(1차원)과 scalar(0차원)의 연산: Broadcasting
print("s3 * 2: ", "\n", s3 * 2)    #series 개별항목에 *scalar

print("\n")

#series와 series의 연산: 동일 인덱스끼리 연산
s4 = pd.Series([1, 2, 3, 4, 5, 6, 7])
print("s3 + s4: ", "\n", s3 + s4)

print("\n")

#사용자정의 함수의 Broadcasting: .apply
s5 = s3.apply(lambda x: 2 * x + 1)     #lambda함수: 일회용 함수
print("사용자정의 함수 연산시키기", "\n", s5)
```

```python
#DataFrame: 행(관측치)과 열(변수)로 이루어진 테이블 데이터 

score_df = pd.DataFrame({
    "KOR": [80, 90, 75],
    "ENG": [90, 80, 70],
    "MATH": [80, 90, 85]
}, index = ["홍길동", "김철수", "이영희"])

display(score_df)
```

```python
#[★]변수의 파생: 기존 변수를 기반으로 새로운 변수 생성

score_df['TOTAL'] = score_df["KOR"] + score_df["ENG"] + score_df["MATH"]
display(score_df)

#새로 만들어진 변수를 기반으로 다시 새로운 변수 생성
score_df['AVERAGE'] = score_df["TOTAL"] / 3
display(score_df)
```

```python
#[★]Boolean 추출

display(score_df[[True, False, True]])
#boolean값을 인덱스로 전달해주면, true를 전달받은 행만 출력

filtered_df = score_df[score_df['AVERAGE'] >= 80]
#socre_df['AVERAGE'] >= 80 연산이 Broadcasting 되는 것.
display(filtered_df)
```

```python
#결측치: 측정되지 않거나 데이터 정제 과정에서 비어져버린 값. 
#        결측치가 보함된 데이터셋의 통계함수 결과는 결측치 

import numpy as np
from numpy import nan, NaN, NAN

exam_scores = pd.Series([90, 80, 120, nan, 95, 80, -10])
print(exam_scores)
```

```python
#결측치 확인하기: pd.isnull()
print(pd.isnull(exam_scores))

print("\n")

#결측빈도 확인하기: np.count_nonzero()
print("결측치의 갯수: ", np.count_nonzero(exam_scores.isnull()))
```

```python
#이상치 확인하기: 
print(~exam_scores.isin(range(0, 101)))  # ~:not

print("\n")

#이상치를 결측치로 대체하기: 
exam_scores[~exam_scores.isin(range(0, 101))] = nan
print(exam_scores)

print("\n")

#이상치를 결측치로 대체한 데이터셋의 통계 구하기: 
print("평균: ", exam_scores.mean())
print("통계량: ", "\n", exam_scores.describe())

print("\n")

#결측치가 많으면 정확한 통계 不  
# --> 결측치를 대표값으로 대체하기:(결측치를 제외한 대푯값)
mid_score = exam_scores[exam_scores.notnull()].median()
print("결측치 제외 중앙값: ", mid_score)

exam_scores[exam_scores.isnull()] = mid_score
print(exam_scores)

print("정제된 데이터의 최종평균: ", exam_scores.mean())
```
