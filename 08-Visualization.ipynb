```python
#시각화의 중요성
#데이터셋이 가진 기초통계량으로만 판단하기에는 무리 (앤스콤의 4분할 그래프)
# --> 데이터 전처리 과정 중 시각화를 통해서 실제 데이터 셋의 모양을 살피는 것이 必

import seaborn as sns

#샘플 데이터셋 확인하기
print(sns.get_dataset_names())
```

```python
#anscombe 데이터셋 불러오기
anscombe = sns.load_dataset('anscombe')
print(type(anscombe))

from IPython.display import display

display(anscombe)
```

```python
#데이터셋 확인하기
display(anscombe.head())
display(anscombe.tail())
print(anscombe.shape)
```

```python
#데이터셋은 dataset 컬럼으로 구분 --총 몇개의 데이터셋이 있는지 확인하기 
print(anscombe['dataset'].nunique())
#.nuique(): 중복제외한 빈도수 
```

```python
#불린연산으로 데이터셋 분할하기
ds1 = anscombe[anscombe['dataset'] == 'I']
ds2 = anscombe[anscombe['dataset'] == 'II']
ds3 = anscombe[anscombe['dataset'] == 'III']
ds4 = anscombe[anscombe['dataset'] == 'IV']
display(ds1, ds2, ds3, ds4)
```

```python
#데이터셋의 기초통계량 확인하기
display(ds1.describe())
display(ds2.describe())
display(ds3.describe())
display(ds4.describe())
```

```python
#ds1의 그래프 그리기
import matplotlib.pyplot as plt

plt.plot(ds1['x'], ds1['y'], 'o')

#마커: o, *, ^, - 
#  -->점그래프(o), 선그래프(-)는 plot함수로 그릴 수 있다! 
# plt.plot(ds1['x'], ds1['y'], '*')
# plt.plot(ds1['x'], ds1['y'], '^')
# plt.plot(ds1['x'], ds1['y'], '-')
```

```python
#앤스콤의 4분할 그래프: 

#기본 틀 생성하기: plt.figure()
fig = plt.figure()

#기본 틀에 격자 추가하기: add_subplot
axe1 = fig.add_subplot(2, 2, 1)     #(2,2)격자 중 1번째 위치
axe2 = fig.add_subplot(2, 2, 2)
axe3 = fig.add_subplot(2, 2, 3)
axe4 = fig.add_subplot(2, 2, 4)

display(fig)

#subplot에 plot 그리기 
axe1.plot(ds1['x'], ds1['y'], 'o', color="green")
axe2.plot(ds2['x'], ds2['y'], '+', color="orange")
axe3.plot(ds3['x'], ds3['y'], '*', color="violet")
axe4.plot(ds4['x'], ds4['y'], '^', color="red")

#subplot의 title 추가하기
axe1.set_title("dataset I")
axe2.set_title("dataset II")
axe3.set_title("dataset III")
axe4.set_title("dataset IV")

#전체title 추가하기
fig.suptitle("Anscombe Dataset")

#레이아웃 조정하기
fig.tight_layout()

#plot 저장하기
plt.savefig("./data/anscombe.png")

#plot 보여주기
plt.show()
```

```python
#Scatter Plot
#mtcars 데이터셋 

import pandas as pd 

mtcars = pd.read_csv("./data/mtcars.csv")

#데이터 형태 확인
print(mtcars.shape)
display(mtcars.head())
display(mtcars.tail())

#기초통계량 확인
display(mtcars.describe())
```

```python
#연비(mpg)와 중량(wt)으로 산점도 그리기
scatter_fig = plt.figure()
scatter_axe = scatter_fig.add_subplot(1, 1, 1)

#산점도 그리기
scatter_axe.scatter(mtcars['wt'], mtcars['mpg'])  #x축: wt, y축: mpg
scatter_axe.set_title("mtcars WT vs MPG")
scatter_axe.set_xlabel("Weight")
scatter_axe.set_ylabel("MPG")

plt.show()
```

```python
#상관계수 구하기: .corr()
mtcars.corr()  #상관계수 표 

#wt와 mpg의 상관계수 구하기:
print("wt와 mpg의 상관계수: ", mtcars.wt.corr(mtcars.mpg))

#상관계수: -1 --역상관관계 
#           0 --상관관계없음
#           1 --정상관관계 
```

```python
#상자그림(Box Plot)  -극단치 제외시키기
#한 변수의 데이터 분포를 시각화 

#mtcars의 연비정보
boxplot = plt.figure()
boxplot_axe = boxplot.add_subplot(1, 1, 1)
boxplot_axe.boxplot([mtcars['mpg'], mtcars['wt']])

plt.show()

#mtcars의 분위수 확인하기 
quart = mtcars.quantile([0.25, 0.5, 0.75])
display(quart)

first_quantile = mtcars['mpg'].quantile(0.25)
third_quantile = mtcars['mpg'].quantile(0.75)
print("mpg의 1Q={}, 3Q={}".format(first_quantile, third_quantile))

#IQR 구하기: 3Q - 1Q
iqr = third_quantile - first_quantile
print("mpg의 IQR: ", iqr)

#상단극단치 경계: 3Q + IQR * 1.5 
#하단극단치 경계: 1Q - IQR * 1.5
top_border = third_quantile + iqr *1.5
bottom_border = first_quantile - iqr * 1.5
print("상단극단치: {}, 하단극단치: {}".format(top_border, bottom_border))

#상단극단치 경계를 넘는 자동차 추출하기: 
mpg_outlier = mtcars[mtcars['mpg'] > top_border]  #boolean 추출
display(mpg_outlier)
```

```python
#히스토그램
#단일변수의 데이터 분포를 구간으로 확인

ws_df = pd.read_csv("./data/wstudents.csv")
display(ws_df)

#히스토그램 그리기: .hist()
#1. figure 만들기
fig = plt.figure()
height_axe = fig.add_subplot(1, 1, 1)

#2. 빈도구간 정의하기
bins = height_axe.hist(ws_df['height'],
                      rwidth = 0.8,
                      bins = range(140, 175, 5))

#3. 표 꾸미기 
height_axe.set_xlabel("Height")
height_axe.set_ylabel("Frequency")
plt.show()

#print(bins)  #빈도표
print("빈도구간: ", bins[1])
print("도수빈도표: ", bins[0])
```

```python
#파이차트
#전체 중의 비율을 시각화

labels = ["A", "B", "C", "D", "F"]
data = [4, 12, 11, 3 , 1]

#figure 만들기
fig = plt.figure()
pie_axe = fig.add_subplot(1, 1, 1)

#파이차트 만들기: 
#1. data 넘겨주기: .pie()
pie_axe.pie(data)

#2. 표 꾸미기
pie_axe.pie(data, 
           labels = labels,      #라벨표시
           autopct = "%.3f%%",   #데이터의 비율 표시
           startangle = 90,      #시작 각도 설정
           explode = (0.2, 0, 0, 0, 0),  #항목별 돌출정도
           shadow = True,
           colors = ["red", "lightblue", "green", "yellow", "gray"]
           ) 

plt.show()
```
